# Parameters and Arguments


```python
#lets take a exapmle in games 

#Difficulty   -easy, medium, hard
#suntitles- trur or false
# paramnetes are difficulty and argumnets are hard or easy

fruits=["apple","lemon","orange",'banana']
weekdays=["monday","tuesday","wednesday","thursday"]

pd.Series(data=fruits)
```

```python
pd.Series(fruits)
```

```python
pd.Series(data=fruits,index=weekdays)  # and remenber index should be match means if their is 
#4 value then other should also have 4 value
```

```python
pd.Series(data=fruits,weekdays)   # shift+Tab will provide some information 
# data is parametes and fruits is a arguments

```

```python
pd.Series(data=fruits,index=weekdays)
```

```python
pd.Series(fruits,index=weekdays)
```

```python
pd.Series(data=fruits,weekdays)  # it will give error
```

```python
fruits=["apple","lemon","orange",'banana',"apple"]
weekdays=["monday","tuesday","wednesday","thursday","monday"]
pd.Series(fruits,weekdays)   # no issue with the duplicates
```
