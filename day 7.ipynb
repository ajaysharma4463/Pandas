# filtering with where method

```python
pokemon=pd.read_excel(r"C:\Users\Ajay\Desktop\pokemon_data.xlsx")
pokemon 
```

# .query methd 

```python
#.query methos can apllied on dataframe
# it doesnot work with column name which has two name like firatname and lastname
pokemon.query("Name=='Bulbasaur'")  # we can prove diferent quotes 
```

```python
pokemon.query("Name !='Bulbasaur'") 
```

```python
pokemon.query("Name !='Bulbasaur' or Name =='Bulbasaur'")    #We can use and ,or  rather than &
```

```python
pokemon.query("Name in ['Bulbasaur','Charmander'] ")
```

```python
pokemon.query("Name not in ['Bulbasaur','Charmander'] ")
```
